```python
from scipy.stats import pearsonr

corr, p_value = pearsonr(new_df['Ghq12'], new_df['Attack'])
print(f"Pearson correlation: {corr:.3f}")
print(f"P-value: {p_value:.4f}")
```

The Pearson correlation of 0.636 indicates a moderate to strong positive relationship between 'Ghq12' and 'Attack'.
As 'Ghq12' (psychological distress) increases, the number of attacks tends to increase as well.
The p-value < 0.001 confirms this relationship is statistically significant

```python
from scipy.stats import ttest_ind

# Separate 'Attack' values by gender
attack_male = new_df[new_df['Gender'] == 'male']['Attack']
attack_female = new_df[new_df['Gender'] == 'female']['Attack']

# Perform t-test
t_stat, p_value = ttest_ind(attack_male, attack_female, equal_var=False)  # Welch's t-test

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.2f}")
```

The t-test shows a t-statistic of -1.99 and a p-value of 0.05, which is right at the common significance threshold. This suggests there is a borderline statistically significant difference in mean 'Attack' counts between males and females. Since the t-statistic is negative, it indicates that males tend to have slightly lower mean attacks than females.

```python
import scipy.stats as stats
import pandas as pd

# Group data by Gender
groups = [group["Attack"].values for name, group in new_df.groupby("Gender")]

# Run one-way ANOVA
f_stat, p_value = stats.f_oneway(*groups)

# Calculate group means rounded to 2 decimals
group_means = new_df.groupby("Gender")["Attack"].mean().round(2)

print(f"ANOVA F-statistic: {f_stat:.2f}")
print(f"P-value: {p_value:.4f}")
print("\nMean Attack by Gender:")
print(group_means)
```

The ANOVA shows a borderline effect of gender on Attack (F = 3.88, p = 0.0513), just above the typical 0.05 significance threshold.
Females have a higher mean Attack count (2.78) compared to males (2.06).
This suggests a trend where gender might influence Attack, but its not quite statistically significant at the 5% level.

```python
from scipy.stats import ttest_ind

# Split 'Attack' by Res_inf groups
attack_yes = new_df[new_df['Res_inf'] == 'yes']['Attack']
attack_no = new_df[new_df['Res_inf'] == 'no']['Attack']

# Run t-test
t_stat, p_value = ttest_ind(attack_yes, attack_no, equal_var=False)  # Welch's t-test

# Calculate group means
mean_yes = attack_yes.mean()
mean_no = attack_no.mean()

print(f"Mean Attack (Res_inf = yes): {mean_yes:.2f}")
print(f"Mean Attack (Res_inf = no): {mean_no:.2f}")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")
```

The mean number of attacks is significantly higher for individuals with 'Res_inf' = yes (3.29) compared to 'no' (1.33). The large t-statistic (6.15) and very small p-value (<0.0001) indicate this difference is statistically significant.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Fit the two-way ANOVA model
model = ols('Attack ~ C(Gender) + C(Res_inf) + C(Gender):C(Res_inf)', data=new_df).fit()

# Perform ANOVA
anova_table = sm.stats.anova_lm(model, typ=2)

# Round results to 2 decimals and display
print(anova_table.round(2))
```

Res_inf has a significant effect on Attack (p < 0.01), while Gender and the interaction between Gender and Res_inf do not show significant effects (p > 0.05). This means Attack varies by Res_inf status but not by Gender or their combination.

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Create contingency table
contingency_table = pd.crosstab(new_df['Gender'], new_df['Res_inf'])

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print results with 2 decimals
print(f"Chi-square statistic: {chi2:.2f}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p:.2f}")
print("\nExpected frequencies:")
print(pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns).round(2))
```

The Chi-square test shows no significant association between Gender and Res_inf ( = 2.18, p = 0.14), so we fail to reject the null hypothesis. This suggests Gender and Res_inf are likely independent in this dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Round values for clarity
new_df['Predicted_Attack'] = new_df['Predicted_Attack'].round(2)
new_df['Ghq12_rounded'] = new_df['Ghq12'].round(2)

# Plot 1: Boxplot of actual vs predicted Attack by Gender
plt.figure(figsize=(8, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x='Gender', y='Attack', data=new_df)
plt.title('Actual Attack by Gender')
plt.xlabel('Gender')
plt.ylabel('Attack Count')

plt.subplot(1, 2, 2)
sns.boxplot(x='Gender', y='Predicted_Attack', data=new_df)
plt.title('Predicted Attack by Gender')
plt.xlabel('Gender')
plt.ylabel('Predicted Attack Count')

plt.tight_layout()
plt.show()

# Plot 2: Lineplot of Ghq12 vs Predicted and Actual Attack
plt.figure(figsize=(8, 4))
sns.scatterplot(x='Ghq12_rounded', y='Attack', data=new_df, label='Actual', alpha=0.6)
sns.lineplot(x='Ghq12_rounded', y='Predicted_Attack', data=new_df, label='Predicted', color='red')
plt.xlabel('Ghq12 (Rounded)')
plt.ylabel('Attack Count')
plt.title('Attack Count vs Ghq12')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns

# Fit Poisson regression model
poisson_model = smf.glm(formula='Attack ~ Gender + Res_inf + Ghq12',
                        data=new_df,
                        family=sm.families.Poisson()).fit()

# Show model coefficients rounded to 2 decimals
summary_table = poisson_model.summary2().tables[1]
print(summary_table.round(2))

# Add predicted values to the dataframe
new_df['Predicted_Attack'] = poisson_model.predict(new_df)

# Plot: Actual vs Predicted Attack
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Predicted_Attack', y='Attack', data=new_df, alpha=0.7)
plt.plot([new_df['Predicted_Attack'].min(), new_df['Predicted_Attack'].max()],
         [new_df['Predicted_Attack'].min(), new_df['Predicted_Attack'].max()],
         color='red', linestyle='--')
plt.xlabel('Predicted Attack')
plt.ylabel('Actual Attack')
plt.title('Actual vs Predicted Attack (Poisson Model)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
