```python
import pandas as pd

# Load the dataset
file_path = "/content/sample_data/INVALSI_data_MAT_2324(Foglio 1).csv"
data = pd.read_csv(file_path)
```

```python
# Show the first 5 rows
print(data.head())
```

```python
# Show variable types
print(" Variable Types:\n")
print(data.dtypes)

# Check completeness (missing values)
print("\nðŸ“Œ Missing Values per Column:\n")
print(data.isnull().sum())

# Quick overview of the dataset
print("\nðŸ“Œ Dataset Info:\n")
print(data.info())
```

```python
selected_columns = [
    "gender", "month", "year", "origin",
    "nursery_attendance", "kindergarten_attendance",
    "father_education", "mother_education",
    "father_occupation", "mother_occupation",
    "Region_Code", "GeoArea_5", "ESCS_student",
    "ESCS_class", "WLE_MATH_200",
    "math_score_raw", "math_score_percentile"
]
data = data[selected_columns].copy()
# Show first rows
print(data.head())

print("\nâœ… Shape of dataset:", data.shape)
```

```python
# Define codes for missing values
missing_codes = {
    "gender": [9],
    "month": [99],
    "year": [9],
    "origin": [9],
    "nursery_attendance": [9],
    "kindergarten_attendance": [9],
    "father_education": [99],
    "mother_education": [99],
    "father_occupation": [99],
    "mother_occupation": [99],
}
# Remove rows with these "missing" codes
for col, codes in missing_codes.items():
    data = data[~data[col].isin(codes)]

# Show results
print("âœ… Cleaned dataset (first 5 rows):")
print(data.head())

print("\nâœ… Shape of cleaned dataset:", data.shape)
```

**modeling**

```python
# Ensure categorical columns are treated as category type
for col in categorical_columns:
    data_clean[col] = data_clean[col].astype('category')
```

```python
import statsmodels.formula.api as smf

# Example: student-level + school-level + contextual-level
# Random intercept for school and region
model = smf.mixedlm(
    "WLE_MATH_200 ~ ESCS_student + C(gender) + C(nursery_attendance) + C(kindergarten_attendance) + ESCS_class + father_education + mother_education + father_occupation + mother_occupation",
    data_clean,
    groups=data_clean["Region_Code"],            # higher-level grouping
    re_formula="~ESCS_class"                     # allows random slopes for school-level variable
)
result = model.fit()
print(result.summary())
```

```python
#--- Extract and show just the variance component for Region ---
region_var = result.cov_re.iloc[0, 0]   # variance of random intercept (region)
region_sd = region_var ** 0.5           # standard deviation

print(f"\nRandom effect (Region variance): {region_var:.3f}")
print(f"Random effect (Region std dev):  {region_sd:.3f}")
```

```python
# --- Extract variance components ---
region_var = result.cov_re.iloc[0, 0]   # variance of region random intercept
residual_var = result.scale              # residual (within-region) variance

# --- Compute ICC ---
icc_region = region_var / (region_var + residual_var)

# --- Display results ---
print(f"Region variance: {region_var:.3f}")
print(f"Residual variance: {residual_var:.3f}")
print(f"ICC (proportion of total variance due to regions): {icc_region:.3f}")
```

**plotting**

```python
import scipy.stats as stats
# Extract model residuals
residuals = result.resid

# Q-Q plot
plt.figure(figsize=(6,6))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title("Q-Q Plot of Mixed-Effects Model Residuals")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.grid(True)
plt.show()
```

```python
# Histogram of residuals
plt.hist(residuals, bins=30, edgecolor='k')
plt.title("Histogram of Residuals")
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.show()
```
